# Вывод

Все пропуски заполнены, хвосты удалены, лишние столбцы тоже. Следующий шаг - подготовка признаков к обучению моделей.

```python
x = data.drop('price', axis=1)
y = data['price']
obj_feat = list(data.loc[:, data.dtypes == 'object'].columns.values)
for feature in obj_feat:
    x[feature] = pd.Series(data[feature], dtype="category")

```

```python
x.info()
```

Хотел сделать стратификацию, однако существуют таргеты, которые встречаются 1 раз. 

```python
x_train, x_valid, y_train, y_valid = train_test_split(x, y, test_size = 0.5, random_state = RS)
x_test, x_valid, y_test, y_valid = train_test_split(x_valid, y_valid, test_size = 0.5, random_state = RS)
```

```python
x_train.shape,x_test.shape,x_valid.shape
```

```python
x.info()
```

Выборки готовы, перейдем к обучению моделей. Всего я обучу 4 модели. CatBoost, LightGBM, линейную регрессию и случайный лес. Помним, что не надо использовать прямое кодирование при обучении CatBoost и LightGBM. Они умеют работать с категориальными признаками.

```python
x_train.info()
```

```python
rmse_scorer = make_scorer(mean_squared_error, squared=False, greater_is_better=False)
```

```python
%%time

cat_сat_features = [0, 2, 4, 6, 7, 8]


model_cat = CatBoostRegressor(cat_features=cat_сat_features)
param_cat_grid = {'learning_rate': np.arange(0.1, 0.5, 0.1),
                 'depth': [ 4, 6, 8, 10],
                 'iterations': [100, 500]}

grid_cat = GridSearchCV(estimator=model_cat, param_grid=param_cat_grid, cv=3, verbose=2, scoring=rmse_scorer)
grid_cat.fit(x_train, y_train)
```

```python
%%time
best_cat = CatBoostRegressor(cat_features=cat_сat_features, **grid_cat.best_params_)
best_cat.fit(x_train, y_train)
```

```python
%%time
y_cat_pred = best_cat.predict(x_valid)
rmse = mean_squared_error(y_valid, y_cat_pred, squared=False)

print("Best parameters: ", grid_cat.best_params_)
print("Best RMSE score: ", rmse)
```

Дальше будем работать с линейной регрессией и случайным лесом. Поэтому применим порядковое кодирование и масштабирование.



```python
x.info()
```

```python
x_ord = data.drop('price', axis=1)
y_ord = data['price']
x_train_ord, x_valid_ord, y_train_ord, y_valid_ord = train_test_split(x_ord, y_ord, test_size = 0.5, random_state = RS)
x_test_ord, x_valid_ord, y_test_ord, y_valid_ord = train_test_split(x_valid_ord, y_valid_ord, test_size = 0.5, random_state = RS)
```

```python
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=RS)
cat_columns = ['vehicle_type','gearbox', 'model', 'fuel_type', 'brand', 'repaired']
encoder.fit(x_train[cat_columns])


x_train_ord[cat_columns] = encoder.transform(x_train_ord[cat_columns])
x_valid_ord[cat_columns] = encoder.transform(x_valid_ord[cat_columns])
x_test_ord[cat_columns] = encoder.transform(x_test_ord[cat_columns])



```

```python
x.info()
```

```python
x_ohe = x.copy()
y_ohe = y.copy()
ohe_features = ['vehicle_type','gearbox', 'model', 'fuel_type', 'brand', 'repaired']
num_features = ['registration_year', 'power', 'kilometer']



x_train_ohe, x_valid_ohe, y_train_ohe, y_valid_ohe = train_test_split(x_ohe, y_ohe, test_size = 0.5, random_state = RS)
x_test_ohe, x_valid_ohe, y_test_ohe, y_valid_ohe = train_test_split(x_valid_ohe, y_valid_ohe, test_size = 0.5, random_state = RS)


ohe = OneHotEncoder(sparse=False, handle_unknown = 'ignore')
ohe_cat_features = ohe.fit(x_train_ohe[ohe_features])

x_train_ohe_ohe = pd.DataFrame(ohe_cat_features.transform(x_train_ohe[ohe_features])).reset_index(drop=True)
x_train_ohe = pd.concat([x_train_ohe_ohe, x_train_ohe[num_features].reset_index(drop=True)], axis=1)

x_valid_ohe_ohe = pd.DataFrame(ohe_cat_features.transform(x_valid_ohe[ohe_features])).reset_index(drop=True)
x_valid_ohe = pd.concat([x_valid_ohe_ohe, x_valid_ohe[num_features].reset_index(drop=True)], axis=1)

x_test_ohe_ohe = pd.DataFrame(ohe_cat_features.transform(x_test_ohe[ohe_features])).reset_index(drop=True)
x_test_ohe = pd.concat([x_test_ohe_ohe, x_test_ohe[num_features].reset_index(drop=True)], axis=1)


x_valid_ohe
```

```python
scaler = StandardScaler()
scaler.fit(x_train_ohe)

x_train_ohe = scaler.transform(x_train_ohe)
x_valid_ohe = scaler.transform(x_valid_ohe)
x_test_ohe = scaler.transform(x_test_ohe)
```

```python
scaler = StandardScaler()
scaler.fit(x_train_ord)

x_train_sca = scaler.transform(x_train_ord)
x_valid_sca = scaler.transform(x_valid_ord)
x_test_sca = scaler.transform(x_test_ord)
```

Дальше перейду к построению моделей.


```python
%%time
model_lin = LinearRegression()
model_lin.fit(x_train_ohe, y_train)

```

```python
%%time
y_lin_pred = model_lin.predict(x_valid_ohe)
rmse_lin = mean_squared_error(y_valid, y_lin_pred, squared=False)


print(" RMSE score линейная регрессия: ", rmse_lin)
```

```python
%%time
model_forest = RandomForestRegressor(random_state=12345)
param_forest_grid = {'max_depth': np.arange(1, 9, 2),
                 'n_estimators':[40,70,110,130,150]}
grid_forest = GridSearchCV(estimator=model_forest, param_grid=param_forest_grid, cv=2, verbose=2,scoring=rmse_scorer)
grid_forest.fit(x_train_sca, y_train)
```

```python
%%time

best_forest = RandomForestRegressor(random_state=12345, **grid_forest.best_params_)
best_forest.fit(x_train_sca, y_train)
```

```python
%%time
y_forest_pred = best_forest.predict(x_valid_sca)
rmse_for = mean_squared_error(y_valid, y_forest_pred, squared=False)

print("Best parameters: ", grid_forest.best_params_)
print("Best RMSE score: ", rmse_for)
```
